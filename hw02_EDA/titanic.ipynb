```python
# анализ пропущенных значений
# isnull() → создаёт таблицу True/False (True там где пропуск)
# .sum()   → суммирует True (True = 1, False = 0) по каждому столбцу
missing = df.isnull().sum()

# Оставляем только столбцы где ЕСТЬ пропуски
missing = missing[missing > 0]

print("\nКоличество пропусков:")
print(missing)

print("\nПроцент пропусков:")
# Делим на общее количество строк и умножаем на 100
missing_percent = (missing / len(df) * 100).round(2)
print(missing_percent)

# Визуализируем пропуски — так нагляднее
plt.figure(figsize=(10, 4))

# Создаём тепловую карту пропусков
# True = пропуск (жёлтый), False = есть данные (фиолетовый)
sns.heatmap(df.isnull(),
# матрица True/False
            yticklabels=False,
# не показываем номера строк
            cbar=False,
# без шкалы цветов
            cmap='viridis')
# цветовая схема

plt.title('Карта пропущенных значений\n(жёлтый = пропуск)')
plt.tight_layout()
plt.show()
```

```python
df_clean = df.copy()
df_clean = df_clean.drop(columns=['Cabin'])
```

```python
print(f"Среднее значение возраста: {df_clean['Age'].mean():.2f}")

# Заполняем средним
df_with_mean = df_clean.copy()
df_with_mean['Age'] = df_with_mean['Age'].fillna(df_with_mean['Age'].mean())
```

на деле заполнение средним на мой взгляд выгляд некорректным. среднее чувствительно к выбросам, реальные возраста могут сильно отличаться, так как распределение очень неравномерное - как видно выше, на титанике были и дети, и старики.

```python
median_age = df_clean['Age'].median()
print(f"Медиана возраста: {median_age}")

df_clean['Age'] = df_clean['Age'].fillna(median_age)
```

медиана уже менее устойчива к выбросам, поэтому вариант лучше. но в реальности видно, что она не сильно отличается от среднего, так что сильно большой разницы бы не было, хоть медиана все-таки надежнее. 

```python
# Embarked — категориальный столбец, пропущено всего 2 значения

print("Значения в Embarked:")
print(df_clean['Embarked'].value_counts())
print()

embarked_mode = df_clean['Embarked'].mode()[0]
df_clean['Embarked'] = df_clean['Embarked'].fillna(embarked_mode)
```

Embarked - категория, поэтому среднее и медиана здесь непреминимы. пропусков всего 2, поэтому при заполнении их модой искажения данных не произойдет. 

```python
# проверка пропусков после очистки
print(df_clean.isnull().sum())
```

```python
# числовые столбцы
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

for col in numeric_cols:
    print(f"{col}:")
    print(f"  min    = {df_clean[col].min():.2f}")
    print(f"  max    = {df_clean[col].max():.2f}")
    print(f"  mean   = {df_clean[col].mean():.2f}")
    print(f"  median = {df_clean[col].median():.2f}")
    print(f"  mode   = {df_clean[col].mode()[0]:.2f}\n")
```

```python

percentiles = [0.05, 0.25, 0.50, 0.75, 0.95]
print(df_clean[['Age', 'Fare']].quantile(percentiles))
```

```python
print("Varience:")
for col in numeric_cols:
    variance = df_clean[col].var()
    std = df_clean[col].std()
# стандартное отклонение = корень из дисперсии
    print(f"  {col}: variance = {variance:.2f}, std = {std:.2f}")
```

```python
print("Skewness:")
for col in numeric_cols:
    skew = df_clean[col].skew()
    if skew > 1:
        direction = "сильный правый хвост"
    elif skew < -1:
        direction = "сильный левый хвост"
    elif skew > 0.5:
        direction = "умеренный правый хвост"
    elif skew < -0.5:
        direction = "умеренный левый хвост"
    else:
        direction = "почти симметричное"
    print(f"  {col}: skewness = {skew:.2f} → {direction}")
```

```python
print("Kurtosis:")
for col in numeric_cols:
    kurt = df_clean[col].kurt()
    if kurt > 1:
        interpretation = "много выбросов, острый пик"
    elif kurt < -1:
        interpretation = "мало выбросов, плоское распределение"
    else:
        interpretation = "близко к нормальному"
    print(f"  {col}: kurtosis = {kurt:.2f} → {interpretation}")
```

```python
from sklearn.feature_extraction import FeatureHasher

df_hash = df_clean.copy()

hasher = FeatureHasher(n_features=8, input_type='string')

# Оборачиваем каждое значение в список
tickets = df_hash['Ticket'].astype(str).apply(lambda x: [x])

hashed_features = hasher.transform(tickets)

hashed_df = pd.DataFrame(
    hashed_features.toarray(),
    columns=[f"Ticket_hash_{i}" for i in range(8)]
)

df_hash = pd.concat([df_hash.reset_index(drop=True), hashed_df], axis=1)

display(df_hash.head())
```
